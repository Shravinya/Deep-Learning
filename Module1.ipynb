1.	Experiment 1: Implementation of a single artificial neuron.



```python
import numpy as np

class SingleNeuron:
    def __init__(self, input_size, learning_rate=0.1):
        # Initialize weights and bias to small random values
        self.weights = np.random.randn(input_size)
        self.bias = np.random.randn()
        self.learning_rate = learning_rate

    # Heaviside step activation function (Binary classification)
    def activation(self, x):
        return 1 if x >= 0 else 0

    # Neuron prediction
    def predict(self, inputs):
        summation = np.dot(inputs, self.weights) + self.bias
        return self.activation(summation)

    # Training the neuron using the Perceptron learning rule
    def train(self, training_inputs, labels, epochs):
        epoch_list = []  # List to store epoch-wise error
        for epoch in range(epochs):
            total_error = 0
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs)
                # Calculate error
                error = label - prediction
                total_error += abs(error)
                # Update weights and bias
                self.weights += self.learning_rate * error * np.array(inputs)
                self.bias += self.learning_rate * error
            epoch_list.append(total_error)
        return epoch_list

# AND gate data (inputs and labels)
training_inputs = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

labels = np.array([0, 0, 0, 1])  # AND gate output

# Create SingleNeuron instance
neuron = SingleNeuron(input_size=2)

# Train the neuron
epochs = 10
epoch_list = neuron.train(training_inputs, labels, epochs)

# Testing the neuron
predictions = [neuron.predict(inputs) for inputs in training_inputs]

# Confusion Matrix calculation
TP = sum((p == 1 and l == 1) for p, l in zip(predictions, labels))
TN = sum((p == 0 and l == 0) for p, l in zip(predictions, labels))
FP = sum((p == 1 and l == 0) for p, l in zip(predictions, labels))
FN = sum((p == 0 and l == 1) for p, l in zip(predictions, labels))

confusion_matrix = [[TP, FP], [FN, TN]]

# Calculate Accuracy
accuracy = (TP + TN) / len(labels)

# Print predictions, confusion matrix, accuracy, and training epochs
print("Predictions:", predictions)
print("Confusion Matrix:")
print(confusion_matrix)
print("Accuracy:", accuracy)
print("Training Epochs:", epochs)
print("Epoch-wise Error:", epoch_list)
```

3.	Experiment 3: Multi-Layer Perceptron (MLP) for XOR Gate.

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

# Define the XOR dataset
X = torch.tensor([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=torch.float32)
y = torch.tensor([[0], [1], [1], [0]], dtype=torch.float32)

# Define the MLP model
class XORMLP(nn.Module):
    def __init__(self):
        super(XORMLP, self).__init__()
        self.hidden = nn.Linear(2, 4)  # Hidden layer with 4 neurons
        self.output = nn.Linear(4, 1)  # Output layer

    def forward(self, x):
        x = F.relu(self.hidden(x))  # Apply ReLU activation to the hidden layer
        x = torch.sigmoid(self.output(x))  # Apply Sigmoid activation to the output layer
        return x

# Initialize the model, loss function, and optimizer
model = XORMLP()
criterion = nn.BCELoss()  # Binary Cross-Entropy Loss
optimizer = optim.SGD(model.parameters(), lr=0.1)

# Training the model
num_epochs = 5000
for epoch in range(num_epochs):
    # Forward pass
    outputs = model(X)
    loss = criterion(outputs, y)

    # Backward pass
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Print loss every 500 epochs
    if (epoch + 1) % 500 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}')

# Testing the model
with torch.no_grad():
    predictions = model(X)
    predictions = (predictions > 0.5).float()  # Convert probabilities to binary output
    print("\nInput:", X.numpy())
    print("Predicted Output:", predictions.numpy())
    print("Actual Output:", y.numpy())
```

5.	Experiment 5: Forward and Backpropagation.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Input data (XOR problem)
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])  # Expected output

# Initialize weights and biases
np.random.seed(42)  # For reproducibility
input_dim = 2
hidden_dim = 2
output_dim = 1

W1 = np.random.uniform(-1, 1, (input_dim, hidden_dim))  # Weights for input to hidden layer
b1 = np.zeros((1, hidden_dim))  # Bias for hidden layer
W2 = np.random.uniform(-1, 1, (hidden_dim, output_dim))  # Weights for hidden to output layer
b2 = np.zeros((1, output_dim))  # Bias for output layer

# Learning rate
lr = 0.1

# Store losses for visualization
losses = []

# Training loop
epochs = 10000
for epoch in range(epochs):
    # Forward propagation
    Z1 = np.dot(X, W1) + b1
    A1 = sigmoid(Z1)  # Hidden layer activation
    Z2 = np.dot(A1, W2) + b2
    A2 = sigmoid(Z2)  # Output layer activation

    # Compute loss (Mean Squared Error)
    loss = np.mean((y - A2) ** 2)
    losses.append(loss)

    # Backpropagation
    dA2 = -(y - A2)  # Loss derivative w.r.t. output
    dZ2 = dA2 * sigmoid_derivative(A2)
    dW2 = np.dot(A1.T, dZ2)
    db2 = np.sum(dZ2, axis=0, keepdims=True)

    dA1 = np.dot(dZ2, W2.T)
    dZ1 = dA1 * sigmoid_derivative(A1)
    dW1 = np.dot(X.T, dZ1)
    db1 = np.sum(dZ1, axis=0, keepdims=True)

    # Update weights and biases
    W2 -= lr * dW2
    b2 -= lr * db2
    W1 -= lr * dW1
    b1 -= lr * db1

    # Print loss every 1000 epochs
    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")

# Test the trained network
print("\nInput:\n", X)
print("Predicted Output:\n", A2)
print("Actual Output:\n", y)

# Visualization of loss over epochs
plt.figure(figsize=(10, 6))
plt.plot(losses, label='Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.show()

# Visualization of decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))

Z1 = sigmoid(np.dot(np.c_[xx.ravel(), yy.ravel()], W1) + b1)
Z2 = sigmoid(np.dot(Z1, W2) + b2)
Z = Z2.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Paired)
plt.scatter(X[:, 0], X[:, 1], c=y.ravel(), edgecolor='k', s=100, cmap=plt.cm.Paired)
plt.title('Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
